2. Tensor Operations

TensorFlow supports various operations that can be performed on tensors, such as element-wise operations, matrix multiplication, reshaping, and more.

```python
import tensorflow as tf

# Define a matrix tensor
matrix_tensor = tf.constant([[1, 2], [3, 4]])
print(matrix_tensor)
ones_tensor = tf.ones_like(matrix_tensor)
print(ones_tensor)
```

```python
# Element-wise addition
tensor_add = tf.add(matrix_tensor, ones_tensor)
print(tensor_add)
matrix_tensor + ones_tensor
```

```python
# Matrix multiplication
matrix_mult = tf.matmul(matrix_tensor, matrix_tensor)
print(matrix_mult)
matrix_tensor @ matrix_tensor
```

```python
#Matrix size-incompatible shape (3, 2)
X = tf.constant([[1, 2], [3, 4], [5, 6]])
Y = tf.constant([[7, 8], [9, 10], [11, 12]])
print(X,Y)
tf.matmul(X, Y)
```

```python
#Reshaping the matrix Y
tf.reshape(Y, shape = (2, 3))
```

```python
#Now matrix multiply using reshaped matrix Y
tf.matmul(X, tf.reshape(Y, shape = (2, 3)))
```

```python
#Changing the datatype of a tensor
B = tf.constant([1.7, 7.4])
C = tf.constant([1, 7])
B, C
```

```python
#Now try to change the datatype of B
B = tf.cast(B, dtype = tf.int32)
B
```

```python
#Tensor Aggregation
#Create a tensor with values between 0 and 100
E = tf.constant(np.random.randint(0, 100, size = 50))
print(E)

#Find the minimum value
E_min = tf.reduce_min(E)
print(E_min)

#Find the maximum value
E_max = tf.reduce_max(E)
print(E_max)

#Find the mean value
E_mean = tf.reduce_mean(E)
print(E_mean)

```

```python
#Squeezing the tensors
G_tensor = tf.constant(np.random.randint(0, 100, size = 50), shape = (1, 1, 1, 1, 50))
print(G_tensor)
G_squeezed = tf.squeeze(G_tensor)
print(G_squeezed)
```

```python
#Create a list of indices
list = [0, 1, 2, 3]
tf.one_hot(list, depth = 4)
```

3. Accessing Elements in a Tensor

You can access specific elements within a tensor using indices. Similar to how you access elements in Python lists or NumPy arrays, TensorFlow provides slicing and indexing operations.

```python
import tensorflow as tf

# Define a vector tensor
vector_tensor = tf.constant([1, 2, 3, 4])

# Accessing the first element of a vector
first_element = vector_tensor[0]

# Define a matrix tensor
matrix_tensor = tf.constant([[1, 2], [3, 4]])

# Slicing a tensor (first two rows of the matrix)
matrix_slice = matrix_tensor[:2, :]
```

4. Changing Tensor Shape

You can change the shape of a tensor by reshaping it. This is often used when you need to feed data into a model with specific input dimensions.

```python
# Reshape a tensor
# changing shape of matrix_tensor to [4, 1]
reshaped_tensor = tf.reshape(matrix_tensor, [4, 1])
print(reshaped_tensor)
```

```python
# Transpose a tensor
# flip rows and columns of matrix_tensor
transpose_tensor = tf.transpose(matrix_tensor)
print(transpose_tensor)
```

**Tensors in Neural Networks**


In neural networks, tensors represent various forms of data throughout the modelâ€™s architecture. For example:

Input data (e.g., images, text, or tabular data) is represented as tensors. For an image, it might be a 4D tensor with the shape [batch_size, height, width, channels].

Weights and biases of layers are represented as tensors. These are the parameters the model learns during training.

Output of layers after applying operations like convolutions, activations, and fully connected layers is also represented as tensors.

How Tensors Flow Through a Neural Network in TensorFlow?

In this example:

The input data X_train is a tensor of shape [100, 10], representing 100 samples with 10 features.

The output of the model is a tensor of shape [100, 1], representing the predictions for each of the 100 samples.

The weights and biases of the model layers are also tensors, and they are updated during training via backpropagation.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Simple dataset
X_train = tf.random.normal([100, 10])
y_train = tf.random.normal([100, 1])
```

```python
# Define the log directory for TensorBoard
log_dir = '/content/drive/My Drive/logs2/'
```

```python
# Define a simple model
modelx = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(32, activation='relu'),
    Dense(1)
])
```

```python
# Compile the model
modelx.compile(optimizer='adam', loss='mse')
```

```python
# Define the TensorBoard callback
tb_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# Train the model with the TensorBoard callback
modelx.fit(X_train, y_train, epochs=5, callbacks=[tb_callback])
```

MNIST DATASET

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive

# Mount Google Drive to save logs
drive.mount('/content/drive')

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Display the number of samples in training and testing datasets
print(f"Number of training samples: {len(x_train)}")
print(f"Number of testing samples: {len(x_test)}")
print(f"Training data shape: {x_train.shape}")

# Define the log directory for TensorBoard
log_dir = '/content/drive/My Drive/logs/'

# Build the model
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')  # Using sigmoid for multi-class classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Define the TensorBoard callback
tb_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

# Train the model with the TensorBoard callback
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test), callbacks=[tb_callback])

# Load the TensorBoard extension
%load_ext tensorboard

# Launch TensorBoard (you can now view it in your Colab environment)
%tensorboard --logdir /content/drive/My\ Drive/logs/
```

### **RANDOM TENSORS**

**Random values from normal distribution**

Generates numbers that can theoretically fall anywhere on the number line (\(-\infty \) to \(+\infty \)).

A pure, uncapped bell curve distribution.

```python
#tf.random_normal is not in the current version
t_random=tf.random.normal([2,3],mean=2.0,stddev=4,seed=0)
t_random
```

Random values from truncated normal distribution

Generates numbers within a fixed range, typically within two standard deviations of the mean. Values outside this range are discarded.

A bell curve with its tails "clipped" or "truncated," resulting in a slightly flatter, squatter shape.

```python
t_truncatedrandom=tf.random.truncated_normal([2,3],mean=2.0,stddev=4,seed=0)
t_truncatedrandom
```
