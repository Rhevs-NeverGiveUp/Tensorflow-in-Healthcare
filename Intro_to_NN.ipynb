# **The basic structure of a neural network**

Artificial Neural Networks (ANNs) are computational models inspired by the human brain's structure.

Neurons: The fundamental building blocks of an ANN are artificial neurons or nodes. Each neuron performs calculations on its input and produces an output.

Layers: Neurons are organized into layers, including an input layer, one or more hidden layers, and an output layer.

Input Layer: Receives the initial data or features.
Hidden Layers: Perform complex computations and extract patterns from the input data.
Output Layer: Generates the network's final prediction or result.

Weights: Each connection between neurons has a numerical value called a weight. Weights determine the strength of the connection and how much influence an input has on a neuron's output.

Biases: Each neuron has a bias term, which is added to the weighted sum of its inputs. Biases allow the neuron to activate even when the weighted sum of inputs is zero, thus enhancing the network's flexibility and learning capacity.

# **Learning from data: a linear regression example**

Neural networks learn from data by adjusting weights and biases iteratively to minimize errors.

Linear Regression is a statistical method for modeling the relationship between a dependent variable and one or more independent variables.

The goal is to find the optimal values for weights and biases that minimize the prediction error, which is often measured by a loss function like Mean Squared Error (MSE).

Gradient Descent is a common optimization algorithm used to adjust weights and biases in the direction that minimizes the loss.

# **Building a fully connected network (FCN) with Keras**

Keras is a high-level neural networks API that simplifies the design and training of deep learning models in TensorFlow.

Sequential API in Keras allows for building models layer-by-layer like a linear stack of layers.

```python
import tensorflow as tf
from tensorflow import keras

# Sample data for demonstration
# Imagine you have a dataset where X is a single feature and y is the target value.
# For example, predicting patient recovery time based on dosage of a medicine.
X_train = tf.constant([[1.0], [2.0], [3.0], [4.0]], dtype=tf.float32)
y_train = tf.constant([[2.0], [4.0], [6.0], [8.0]], dtype=tf.float32)

# Build a simple Fully Connected Network (FCN) using Keras Sequential API
model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1]) # A single dense layer for linear regression
])

# Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error') # Stochastic Gradient Descent as optimizer, MSE as loss

# Train the model
model.fit(X_train, y_train, epochs=100) # Train for 100 epochs

# Make a prediction
print(model.predict(tf.constant([[5.0]], dtype=tf.float32)))
```

In the above example:

The code demonstrates building a simple linear regression model (a form of FCN) using the Keras Sequential API.

The Dense layer represents the fully connected nature of the network, where each neuron in the layer is connected to all neurons in the previous layer.

The optimizer (sgd - Stochastic Gradient Descent) is responsible for adjusting the weights and biases during training to minimize the loss.

The loss function (mean_squared_error) calculates the difference between the model's predictions and the actual values, guiding the optimization process.

The epochs parameter specifies how many times the model will iterate over the entire training dataset during the training process.

This basic example illustrates how Keras simplifies the process of building and training neural networks for learning from data, even for a fundamental task like linear regression, which lays the groundwork for understanding more complex deep learning architectures in healthcare.
